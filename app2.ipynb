```python
import pyautogui
import keyboard
import time

def processar_cnpj(cnpj):
    time.sleep(10)
    #------------------------------------------1-----------------------
    pyautogui.press('tab', presses=17)
    time.sleep(3)
    
    pyautogui.press('enter', presses=1)
    time.sleep(5)
    
    pyautogui.press('tab', presses=2)
    time.sleep(5)
    
    pyautogui.write(cnpj)
    time.sleep(3)
    
    pyautogui.press('tab', presses=1)
    time.sleep(3)
    
    pyautogui.press('enter', presses=1)
    time.sleep(5)

# Lista de CNPJs para processar
cnpjs = ['48951224000105', '12816158000113', '45451221000114']
```

```python
import pyautogui
import keyboard
import time
# Loop para processar cada CNPJ
for cnpj in cnpjs:
    processar_cnpj(cnpj)

declaracao = pyautogui.locateCenterOnScreen('bnt_declaracao.png')
time.sleep(3)

pyautogui.click(declaracao[0],declaracao[1])
time.sleep(3)

pyautogui.press('tab', presses=10)
time.sleep(3)

pyautogui.press('enter', presses=1)
time.sleep(3)
```
